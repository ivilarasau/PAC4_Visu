```python
# Importem les llibreries necessàries per la realització del projecte: 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Carreguem les dades de la pax a un dataframe de pandas:
df2 = pd.read_csv("pax_data_1789_agreements_17-11-19.csv")
# Mostrem per pantalla els atributs del nou dataframe i les seves dimensions:
print(df2.head(n=0))
print(df2.shape)
```

```python
# En aquesta cel·la eliminarem del dataframe totes les columnes que continguin valors na:
nan_cols = [i for i in df2.columns if df2[i].isna().any()]
df2.dropna()
print(nan_cols)
```

```python
# En aquest punt agafarem les dades dels 6 països de mostra (explicació feta en el document entregable) per tal d'estudiar 
# quins atributs tenen més correlació entre aquests països per més tard poder-ho extrapolar a la resta de països 
#dels acords de pau:
list_ = ['Somalia/Puntland', 'Indonesia/Aceh', 'Papua New Guinea/Bougainville', 'Yemen', 'Nepal', 'Philippines/Mindanao']
con_sample = df2[df2['Con'].isin(list_)]
#Papua = df2[df2['Con']=='Papua New Guinea/Bougainville']
#print(Papua)
```

```python
# Aquí creem nous dataframes amb un nombre reduït d'atributs per tal de poder estudiar les correlacions entre ells
# mitjançant visualitzacions de mapes de colors més intel·ligibles:

corr = con_sample.iloc[:, 0:50].corr()
corr2 = con_sample.iloc[:, 51:100].corr()
corr3 = con_sample.iloc[:, 101:150].corr()
corr4 = con_sample.iloc[:, 151:200].corr()
corr5 = con_sample.iloc[:, 201:250].corr()
corr6 = con_sample.iloc[:, 251:266].corr()

# Mostrarem per pantalla tots els mapes de colors:

plt.figure(figsize=(10,10))
plt.title('Corr1')
sns.heatmap(corr,
            vmin=-1,
            cmap='coolwarm',
            annot=False);

plt.figure(figsize=(10,10))
plt.title('Corr2')
sns.heatmap(corr2,
            vmin=-1,
            cmap='coolwarm',
            annot=False);

plt.figure(figsize=(10,10))
plt.title('Corr3')
sns.heatmap(corr3,
            vmin=-1,
            cmap='coolwarm',
            annot=False);

plt.figure(figsize=(10,10))
plt.title('Corr4')
sns.heatmap(corr4,
            vmin=-1,
            cmap='coolwarm',
            annot=False);

plt.figure(figsize=(10,10))
plt.title('Corr5')
sns.heatmap(corr5,
            vmin=-1,
            cmap='coolwarm',
            annot=False);

plt.figure(figsize=(10,10))
plt.title('Corr6')
sns.heatmap(corr6,
            vmin=-1,
            cmap='coolwarm',
            annot=False);
```

```python
# A partir dels atributs que hem vist anteriorment que disposen d'una correlació més alta, hem creat aquest nou
# dataframe per tal d'estudiar amb detall quins són els atributs amb més correlació dins dels països de mostra:

corr7 = con_sample.iloc[:, 100:200].corr()
# Mostrem per pantalla el mapa de colors de correlacions:
plt.figure(figsize=(20,20))
plt.title('Corr')
sns.heatmap(corr7,
            vmin=-1,
            cmap='coolwarm',
            annot=False);
```
